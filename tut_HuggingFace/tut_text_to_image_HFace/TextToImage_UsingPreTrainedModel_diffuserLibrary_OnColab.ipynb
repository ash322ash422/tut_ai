```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model (requires a Hugging Face account login)
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Check if CUDA is available before moving the model to GPU
if torch.cuda.is_available():
    pipe = pipe.to("cuda")  # Use GPU for faster generation, if available. It took 1 minute
else:
    print("Warning: CUDA is not available. Using CPU instead.") #Was taking loooong time

# Set your text prompt
prompt = "A huge red balloon flying outside the city."

# Generate the image
image = pipe(prompt).images[0]

# Save or display the image
image.save("generated_image.png")
image.show()  # To display in Jupyter or other Python environments
```
